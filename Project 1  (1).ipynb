```python
labels = data['Age'].head(5)
sizes= data['ProductCategory'].head(5)
plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title("Pie chart For Age and ProductCategory")
plt.show()
```

```python
Age_data=data['Age'].head(20)
NOP = data['ProductCategory'].head(20)
index= data['NumberOfPurchases'].head()
df = pd.DataFrame({'Age':Age_data,'ProductCategory': NOP }, index=index)
plot = df.plot.pie(y='Age', figsize=(5, 5))
```

```python
dt=data[data.index.isin(range(2,50))]
NOP = dt['NumberOfPurchases'].head(10)
DA = dt['DiscountsAvailed'].head(10)

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(NOP, DA)
plt.subplot(132)
plt.scatter(NOP, DA)
plt.subplot(133)
plt.plot(NOP, DA)
plt.suptitle('Categorical Plotting btw NumberOfPurchases and DiscountsAvailed')
plt.show()
```
