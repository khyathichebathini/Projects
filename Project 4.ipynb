```python
import matplotlib.pyplot as plt
labels = data['Title'].head(5)
sizes= data['IMDb Rating'].head(5)
plt.pie(sizes, labels=labels, autopct="%1.0f%%", startangle=100)
plt.title("Pie chart For Title and IMDb Rating")
plt.show()
```

```python
import matplotlib.pyplot as plt
labels = data['Genre'].head(5)
sizes= data['MetaScore'].head(5)
plt.pie(sizes, labels=labels, autopct="%1.0f%%", startangle=100)
plt.title("Pie chart For Genre and MetaScore")
plt.show()
```

```python
dt=data[data.index.isin(range(10,20))]
dt1=dt.loc[dt['Duration (minutes)']>= 90]
sns.barplot(data=dt1 ,x='Duration (minutes)',y='MetaScore')
```
