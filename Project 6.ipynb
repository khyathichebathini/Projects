```python
labels = data['Vict Sex'].head(10)
sizes= data['Crm Cd'].head(10)
plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title("Pie chart For Vict Sex and Weapon Desc	")
plt.show()
```

```python
sns.countplot(x='Vict Sex', data=data)
plt.title('Gender Distribution of Victims')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
sns.histplot(data['Vict Age'].dropna(), bins=30, kde=True)
plt.title('Distribution of Victim Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
